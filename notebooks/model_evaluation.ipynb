```python
import joblib
import pandas as pd
from sklearn.metrics import classification_report, precision_recall_curve
import matplotlib.pyplot as plt
```

```python
# Load preprocessed dataset
df = pd.read_csv(r"C:\Users\Anilchoudary R\Gitrepos\bank-API-1\data\processed_data\creditcard.csv")
```

```python
X = df.drop(columns=["Class"])
y = df["Class"]
```

```python
# Load trained model
model = joblib.load(r"C:\Users\Anilchoudary R\Gitrepos\bank-API-1\models\creditcard_model_xgb.pkl")
```

```python
y_pred = model.predict(X)
```

```python
print(classification_report(y, y_pred))
```

```python
precision, recall, _ = precision_recall_curve(y, model.predict_proba(X)[:, 1])

# Plot Precision-Recall Curve
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker=".", label="XGBoost")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.legend()
plt.show()
```
